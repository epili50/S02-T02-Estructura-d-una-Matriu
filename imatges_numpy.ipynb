### Ejercicio 3

```python
arrBidem = np.random.randint(100, size=(5, 5)) 

print(arrBidem)

valorMax = np.max(arrBidem)

print("El número más grande de la matriz es:", valorMax)

#ver diferencias entre max y amax

numMaxFila = np.amax(arrBidem, axis = 1)
print("El número máximo de cada fila es", numMaxFila)

numMaxColumna = np.amax(arrBidem, axis = 0)
print("El número máximo de columna fila es", numMaxColumna)
```

### Ejercicio 5
Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.

```python
ej = np.random.randint(100, size=(5, 5)) 
print(ej)

c = 1 #columna que queremos obtener
print ("Número de la colmna de index 1 ",ej[:,c])
f = 3 #fila que queremos obtener
print("Número de la fila index 3 ",ej[f])

#suma de la fila mas la columna

ej[:,c] + ej[f]
```
